## Understanding about the columns

## Column Name  	          Description 
### Area_Name	             :Name of the state where data is recorded.
### Year	                 :The reporting year of the crime data.
### Group_Name	             :Main category of crime reported.
### Sub_Group_Name	         :Specific type of crime under the main group.
### PT_1_3_Years	         :Cases pending trial for 1 to 3 years.
### PT_3_5_Years	         :Cases pending trial for 3 to 5 years.
### PT_5_10_Years	         :Cases pending trial for 5 to 10 years.
### PT_6_12_Months	         :Cases pending trial for 6 to 12 months.
### PT_Less_than_6_Months	 :Cases pending trial for less than 6 months.
### PT_Over_10_Years	     :Cases pending trial for more than 10 years.
### PT_Total	             :Total number of pending trial cases across all durations.

```python
pt_columns = [
    'PT_1_3_Years',
    'PT_3_5_Years',
    'PT_5_10_Years',
    'PT_6_12_Months',
    'PT_Less_than_6_Months',
    'PT_Over_10_Years'
]

df2[pt_columns] = df2[pt_columns].fillna(0)

df2['PT_Total'] = df2[pt_columns].sum(axis=1)
```

```python
print("\nMissing values after cleaning:")
print(df2.isnull().sum())
```

## Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

## Univariate Analysis

### Distribution Of Years

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Year', data=df2, palette='crest')
plt.title("Distribution of Records by Year")
plt.xticks(rotation=45)
plt.show()
```

### Top 10 Areas By Total Cases

```python
top_areas = df2.groupby('Area_Name')['PT_Total'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_areas.values, y=top_areas.index, palette='viridis')
plt.title("Top 10 Areas by Total Cases")
plt.xlabel("Total Cases")
plt.ylabel("Area Name")
plt.show()
```

### Distribution Of Trial Duration

```python
pt_cols = [
    'PT_Less_than_6_Months', 'PT_6_12_Months', 'PT_1_3_Years',
    'PT_3_5_Years', 'PT_5_10_Years', 'PT_Over_10_Years'
]

df2[pt_cols].sum().plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title("Total Number of Cases by Trial Duration")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.show()
```

## Bivariant Analysis

### Year VS Toal Cases

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='PT_Total', data=df2, estimator='sum', ci=None, marker='o')
plt.title("Trend of Total Cases Over Years")
plt.xlabel("Year")
plt.ylabel("Total Cases")
plt.grid(True)
plt.show()
```

### Group wise Total Cases

```python
group_total = df2.groupby('Group_Name')['PT_Total'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=group_total.values, y=group_total.index, palette='mako')
plt.title("Total Cases by Group Name")
plt.xlabel("Total Cases")
plt.ylabel("Group Name")
plt.show()
```

### Correlation Analysis

```python
plt.figure(figsize=(8,6))
sns.heatmap(df2[pt_cols + ['PT_Total']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Trial Durations")
plt.show()
```

### Outlier Detection

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df2[pt_cols])
plt.title("Outlier Check for Trial Duration Columns")
plt.xticks(rotation=45)
plt.show()
```

```python
df2['Long_Duration'] = df2['PT_5_10_Years'] + df2['PT_Over_10_Years']
df2['Short_Duration'] = df2['PT_Less_than_6_Months'] + df2['PT_6_12_Months']

df2['Long_Ratio'] = df2['Long_Duration'] / df2['PT_Total']

top_long = df2.groupby('Area_Name')['Long_Ratio'].mean().sort_values(ascending=False).head(10)
top_long.plot(kind='bar', figsize=(10,5), title='Top 10 Areas with Highest Long-duration Case Ratio')
plt.show()
```

### Top 10 Areas with Highest Short-duration Case Ratio

```python
df2['Short_Duration'] = df2['PT_Less_than_6_Months'] + df2['PT_6_12_Months']

df2['Short_Ratio'] = df2['Short_Duration'] / df2['PT_Total']

top_long = df2.groupby('Area_Name')['Short_Ratio'].mean().sort_values(ascending=False).head(10)
top_long.plot(kind='bar', figsize=(10,5), title='Top 10 Areas with Highest Short-duration Case Ratio')
plt.show()
```

### Sub-Group level Total Cases
